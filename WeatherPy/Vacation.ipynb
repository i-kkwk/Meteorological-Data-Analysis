### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#configure gmaps with g_key
gmaps.configure(g_key)
```

```python
#Store "Latitude" and "Longtitude" data into locations, "Humidity" data into humid_rates and convert to float.

locations = cities_df[["Lat","Lng"]].astype(float)

humid_rates = cities_df ["Humidity"].astype(float)
```

```python

#Plot Heatmap

fig = gmaps.figure()

#Get heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humid_rates, 
                                 dissipating = False, max_intensity=50, point_radius=3)
#Add layer
fig.add_layer(heat_layer)

#Display
fig
```

### Create new DataFrame fitting weather criteria
